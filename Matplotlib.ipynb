# Example

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
# exponensial
```

```python
plt.plot(x, y, 'r') #'r' warna merah
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

# Multiplots on Same Canvas

```python
# plt.subplot(nbaris, nkolom, nomor_plot)
plt.subplot(1, 2, 1)
# '--' bentuk grafik
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
# '*-' bentuk grafik
plt.plot(y, x, 'g*-');
```

# Object Oriented Method (OOM)

```python
# empty canvas
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(x, y, 'b')
axes.set_xlabel('Set x Label') #set_
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
# Blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # Main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # Insert axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

# Subplots()

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# index baris/kolom mulai dari 0
```

```python
axes
```

```python
# Iterasi
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
# Display figure object
fig
```

```python
# Overlapping/Numpuk-numpuk
# fig.tight_layout()/plt.tight)layout()
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig
plt.tight_layout()
```

```python
# Legends
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label='x**2')
ax.plot(x, x**3, label='x**3')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
```

```python
# Option 1
ax.legend(loc=1) # Upper right corner
ax.legend(loc=2) # Upper left corner
ax.legend(loc=3) # Lower left corner
ax.legend(loc=4) # Lower right corner

# Option 2 (Most Common)
ax.legend(loc=0) # Matplotlib menentukan lokasi optimal
fig
```

# Setting colors, Linewidths, Linetypes

```python
# Colors with MatLab like syntax
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b--') # Blue line w/ dots
ax.plot(x, x**3, 'g--') # Green dashed line
```

```python
# Colors w/ the color = parameter
```

```python
# linewidths/lw
fig, axes = plt.subplots(nrows=1, ncols=1)

axes.plot(x, -y, color='#8B008B', lw=10, label='Performa') # RGB colors
axes.set_xlabel('Usia')
axes.set_ylabel('Performa')
axes.set_title('Usia dan Performa')
axes.legend()
```

# Plot Range

```python
fig, axes = plt.subplots(nrows=1, ncols=1)

axes.plot(x, -y, 'r--', lw=10, label='Performa')
axes.set_xlabel('Usia')
axes.set_ylabel('Performa')
axes.set_title('Usia dan Performa')
axes.set_xlim(0, 50)
axes.set_ylim(-100, 0)
axes.legend()
```
